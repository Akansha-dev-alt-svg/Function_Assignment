THEORY QUESTIONS WITH EXAMPLES

Answer_1: Function vs Method

 Function: Independent block of code.

 Method: Function associated with an object.

    def greet():
        return "Hello"

    print(greet())  # Function

    s = "hello"
    print(s.upper())  # Method

Answer_2:  Arguments vs Parameters

  Parameters: Inside function definition.

  Arguments: Actual values passed.

    def add(a, b):  # a, b are parameters
        return a + b

    add(2, 3)  # 2, 3 are arguments

Answer_3:  Defining and Calling Functions

    Normal function
    def square(x):
        return x*x
    print(square(5))

    Lambda
    square = lambda x: x*x
    print(square(5))

Answer_4:  Return Statement

Used to return a result from a function

    def get_sum(a, b):
        return a + b
    print(get_sum(10, 5))

Answer_5: Iterators vs Iterables

Iterable: Can be looped (like list)

Iterator: Object with __next__()

    nums = [1, 2, 3]
    iterator = iter(nums)
    print(next(iterator))

Answer_6: Generators

Use yield, pause after each yield

    def gen_nums():
        yield 1
        yield 2
    for i in gen_nums():
        print(i)

Answer_7: Advantages of Generators

Memory efficient, lazy evaluation

    def gen_squares():
        for i in range(5):
            yield i*i

    for val in gen_squares():
        print(val)

Answere_8: Lambda Function

Small anonymous function

    add = lambda x, y: x + y
    print(add(3, 4))

Answer_9: map() Function

Applies function to all items

    nums = [1, 2, 3, 4]
    squares = list(map(lambda x: x*x, nums))
    print(squares)

Answer_ 10: map vs reduce vs filter

from functools import reduce

# map: Applies function to all
    print(list(map(lambda x: x*2, [1, 2, 3])))
# reduce: Reduces to single value
    print(reduce(lambda x, y: x+y, [1, 2, 3]))
# filter: Filters values
    print(list(filter(lambda x: x%2 == 0, [1, 2, 3, 4])))

# PRACTICAL QUESTIONS

```python
# 1. Sum of even numbers

def sum_even(lst):
    return sum([x for x in lst if x % 2 == 0])
print(sum_even([1, 2, 3, 4]))
```

```python
# 2. Reverse a string

def reverse_str(s):
    return s[::-1]
print(reverse_str("hello"))
```

```python
# 3. Squares of list numbers

def square_list(lst):
    return [x*x for x in lst]
print(square_list([1, 2, 3]))
```

```python
# 4. Check prime up to 200

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

for i in range(1, 201):
    if is_prime(i):
        print(i, end=" ")
print()
```

```python
# 5. Fibonacci Iterator

class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return val

for num in Fibonacci(7):
    print(num, end=" ")
print()
```

```python
# 6. Powers of 2 Generator

def powers_of_2(exp):
    for i in range(exp + 1):
        yield 2 ** i

for power in powers_of_2(5):
    print(power)
```

```python
# 7. File line generator

def read_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()
# Example usage:
# for line in read_lines('file.txt'):
#     print(line)
```

```python
# 8. Sort tuples by second value

tuple_list = [(1, 3), (2, 1), (4, 2)]
sorted_list = sorted(tuple_list, key=lambda x: x[1])
print(sorted_list)
```

```python
# 9. Convert Celsius to Fahrenheit using map

def c_to_f(c):
    return (c * 9/5) + 32

celsius = [0, 20, 37, 100]
fahrenheit = list(map(c_to_f, celsius))
print(fahrenheit)
```

```python
# 10. Remove vowels using filter

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda x: x not in vowels, s))

print(remove_vowels("Beautiful Day"))
```

```python
# 11. List of orders: [order number, book title, quantity, price per item]
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99]
]

# Use lambda and map to compute the required result
result = list(map(lambda order: (
    order[0],
    order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10
), orders))

# Output the result
print(result)
```
